```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df_F1 = pd.read_excel('results_F1.xlsx')
df_recall = pd.read_excel('results_recall.xlsx')
df_precision = pd.read_excel('results_precision.xlsx')

colors = ['#1F77B4', '#AEC7E8', '#FF7F0E', '#FFBB78', '#2CA02C', '#98DF8A', '#D62728', '#FF9896', '#9467BD', '#C5B0D5',
          '#8C564B', '#C49C94', '#E377C2', '#F7B6D2', '#7F7F7F',  '#C7C7C7', '#BCBD22', '#DBDB8D', '#17BECF', '#9EDAE5']
```

```python
df_video = pd.read_excel('results_collection.xlsx')
```

```python
df_video = df_video.transpose().iloc[1:, [2, 3, 5, 7]]
```

```python
df_F1= df_F1.set_index('Lecture name')
```

```python
df_video = df_video.rename(columns={
    2: 'image quality',
    3: 'sound quality',
    5: 'jumpiness',
    7: 'nonslideratio',
})
```

```python
# Join df1 and df2
result = df_video.join(df_F1)
result = result.iloc[:,:]
result
```

```python
# Set Seaborn style
sns.set_style("white")


corr_image = result['All 0.1 mean'].corr(result['image quality'], method='spearman')
print(corr_image)

corr_sound = result['All 0.1 mean'].corr(result['sound quality'], method='spearman')
print(corr_sound)

#plt.scatter(result['All 0.1 mean'], result['jumpiness'])
# plt.scatter(result['All 0.1 mean'], result['image quality'], label='Image quality, Correlation={}'.format(corr_image.round(2)), color=colors[0] , s=40)
# plt.scatter(result['All 0.1 mean'], result['sound quality'], label='Sound quality, Correlation={}'.format(corr_sound.round(2)), color=colors[2],  s=40)

sns.regplot(x='All 0.1 mean', y='image quality', data=result,
            color=colors[0],  # Color of the data points
            line_kws={'color': colors[0], 'label': 'Image quality, r={}'.format(corr_image.round(2))},  # Properties of the regression line
            ci=None)  # Confidence interval size, set to None for no CI

sns.regplot(x='All 0.1 mean', y='sound quality', data=result,
            color=colors[2],  # Color of the data points
            line_kws={'color': colors[2], 'label': 'Sound quality, r={}'.format(corr_sound.round(2))},  # Properties of the regression line
            ci=None)  # Confidence interval size, set to None for no CI

plt.legend()

plt.ylabel('Likert scale: 1=inferior to 7=very good', fontsize=14)
plt.legend(loc='upper right', fontsize=14)

plt.ylim([1, 7])

plt.xlabel('F1 score', fontsize=14)

plt.xticks([0.4, 0.6, 0.8, 1])

sns.despine()
plt.rc('font', family='serif', size=14)
plt.savefig('image_sound_quality_corr.pdf', format='pdf') # For PDF format
# or
plt.savefig('image_sound_quality_corr.svg', format='svg') # For SVG format

plt.show()
```

```python
# Set Seaborn style
from matplotlib.lines import Line2D

sns.set_style("white")

plt.figure(figsize=(7, 7))


corr_image = result['Audio 0.1'].corr(result['image quality'], method='spearman')
print(corr_image)

corr_sound = result['Audio 0.1'].corr(result['sound quality'], method='spearman')
print(corr_sound)

corr_image_ocr = result['OCR 0.1'].corr(result['image quality'], method='spearman')
print(corr_image_ocr)

corr_sound_ocr = result['OCR 0.1'].corr(result['sound quality'], method='spearman')
print(corr_sound_ocr)

sns.regplot(x='Audio 0.1', y='image quality', data=result.astype(float),
            color=colors[0],  # Color of the data points,
            marker= "^",
            scatter_kws={"s": 80},
            line_kws={'color': colors[0], 'label': 'Image, r={}'.format(corr_image.round(2))},  # Properties of the regression line
            ci=95)  # Confidence interval size, set to None for no CI

sns.regplot(x='Audio 0.1', y='sound quality', data=result.astype(float),
            color=colors[2],  # Color of the data points
            marker='x',
            scatter_kws={"s": 80},
            line_kws={'color': colors[2], 'label': 'Sound, r={}'.format(corr_sound.round(2)), 'linestyle':'dashed'},  # Properties of the regression line
            ci=95)  # Confidence interval size, set to None for no CI

# After plotting your regplots, create custom legend handles

legend_elements = [
    Line2D([0], [0], marker='^', color=colors[0], markerfacecolor=colors[0], markersize=10,
           label='Image, r={}'.format(round(corr_image, 2))),
    Line2D([0], [0], marker='x', color=colors[2], markerfacecolor=colors[2], markersize=10,
           label='Audio, r=0.60', linestyle='dashed')
]

# Create the legend
plt.legend(handles=legend_elements, fontsize=30, loc='upper center', bbox_to_anchor=(0.5, -0.15),
           shadow=True, ncol=1, frameon=False, handlelength=2, handletextpad=0.5, columnspacing=1)

plt.xlabel('F1 score')
plt.ylabel('Image / Audio quality', fontsize=30)

# plt.legend(loc='upper right', fontsize=16)

plt.xlabel('F1 score', fontsize=30)

sns.despine()
plt.rc('font', family='serif', size=30)

plt.xticks([0, 0.2, 0.4, 0.6, 0.8, 1], fontsize=30)
plt.yticks(fontsize=30)

plt.ylim([0.5, 7.5])


plt.yticks([1,2,3,4,5,6,7], fontsize=30)

sns.despine()

plt.savefig('image_sound_quality_corr_audio.pdf', format='pdf', bbox_inches='tight') # For PDF format
# or
plt.savefig('image_sound_quality_corr_audio.svg', format='svg', bbox_inches='tight') # For SVG format

plt.show()
```

```python
# Set Seaborn style
sns.set_style("white")

plt.figure(figsize=(7, 7))


plt.rcParams["font.serif"] = "Times New Roman NN"

result['jumpiness'] = result['jumpiness'].astype(float)
corr_jump0 = result['All 0.0 max'].corr(result['jumpiness'], method='spearman')
corr_jump1 = result['All 0.1 max'].corr(result['jumpiness'], method='spearman')
corr_jump15 = result['All 0.15 max'].corr(result['jumpiness'], method='spearman')
corr_jump2 = result['All 0.2 max'].corr(result['jumpiness'], method='spearman')
corr_jump25 = result['All 0.25 max'].corr(result['jumpiness'], method='spearman')

sns.regplot(x='All 0.0 max', y='jumpiness', data=result.astype(float),
            color=colors[0],  # Color of the data points
            marker= 'x',
            scatter_kws={"s": 80},
            line_kws={'color': colors[0], 'label': '$\lambda_{jump}$=0,' + ' r=${}$,'.format(corr_jump0.round(2))},  # Properties of the regression line
            ci=95)  # Confidence interval size, set to None for no CI


sns.regplot(x='All 0.1 max', y='jumpiness', data=result.astype(float),
            color=colors[2],  # Color of the data points
            marker= '^',
            scatter_kws={"s": 80},
            line_kws={'color': colors[2], 'label': '$\lambda_{jump}$=0.1,' + ' r=${}$'.format(corr_jump1.round(2)), 'linestyle':'--'},  # Properties of the regression line
            ci=95)  # Confidence interval size, set to None for no CI


sns.regplot(x='All 0.2 max', y='jumpiness', data=result.astype(float),
            color=colors[4],  # Color of the data points
            scatter_kws={"s": 80},
            marker='o',
            line_kws={'color': colors[4], 'label': '$\lambda_{jump}$=0.2,' + ' r=${}$'.format(corr_jump2.round(2)), 'linestyle':':'},  # Properties of the regression line
            ci=95)  # Confidence interval size, set to None for no CI


legend_elements = [
    Line2D([0], [0], marker='x', color=colors[0], markerfacecolor=colors[0], markersize=10,
           linestyle='solid', label='$\lambda^{jump}=0.0$,' + ' $r={}$'.format(corr_jump0.round(2))),
    Line2D([0], [0], marker='^', color=colors[2], markerfacecolor=colors[2], markersize=10,
           linestyle='--', label='$\lambda^{jump}=0.1$,' + ' $r={}$'.format(corr_jump15.round(2))),
    Line2D([0], [0], marker='o', color=colors[4], markerfacecolor=colors[4], markersize=10,
           linestyle=':', label='$\lambda^{jump}=0.2$,' + ' $r={}$'.format(corr_jump2.round(2)))
]

# Create the legend
plt.legend(handles=legend_elements, fontsize=30, loc='upper center', bbox_to_anchor=(0.5, -0.15),
           shadow=True, ncol=1, frameon=False, handlelength=2, handletextpad=0.5, columnspacing=1)

plt.xlabel('F1 score', fontsize=30)
plt.ylabel('volatility score', fontsize=30)
plt.xticks([0.4,0.6, 0.8, 1])
plt.yticks([0, 1, 2, 3, 4, 5])

plt.xlabel('F1 score', fontsize = 30)
plt.rcParams['xtick.labelsize'] = 30
plt.rcParams['ytick.labelsize'] = 30

sns.despine()

plt.xticks( fontsize=30)
plt.yticks( fontsize=30)

sns.despine()
plt.rc('font', family='serif', size=30)
plt.savefig('volatility_jump_penalty.pdf', format='pdf', bbox_inches='tight') # For PDF format
# or
plt.savefig('volatility_jump_penalty.svg', bbox_inches='tight') # For SVG format

plt.show()
```
